```python
dataset_dir = "../Datasets/"
data_file = "COVID-19 Cases(14-03-2022).csv"
```

```python
dat = pd.read_csv(dataset_dir + data_file)
dat = dat[[int(x[-4:]) >= 2020 for x in dat["Date"]]]
dat = dat[dat["Region"] != "World"]
dat.columns

#create a p_death column
dat["p_d_t"] = dat['Death'] / (dat['Cured/Discharged'] + dat['Death'])
```

```python
dat.Date =  [pd.Timestamp(year = int(x[2]), month = int(x[1]), day = int(x[0])) for x in dat.Date.str.split("/")]
```

```python
dat.columns
```

```python
c_max = dict(dat[["Date", "Region", "Confirmed Cases"]].pivot_table('Confirmed Cases', ['Date'], 'Region').fillna(0).max())
```

```python
dat["weight_t"] = dat["Active Cases"] / dat["Region"].replace(c_max)
```

```python
dat["p_d_w_t"] = dat.p_d_t * dat.weight_t
```

```python
dat
```

```python
d_trans = dat[["Date", "Region", "p_d_w_t"]].pivot_table('p_d_w_t', ['Date'], 'Region').fillna(0)
fig = d_trans.plot(subplots = True, sharex = True, ylim = (0,0.005))
```
